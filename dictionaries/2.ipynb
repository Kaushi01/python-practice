```python
d= {"name": ["alice", "picachu" ,"wow" ],
    "age":  [90, 70, 80],
    "id": [101, 201, 301],
    "marks": [89, 90, 87]}

d.pop("marks")
del d["age"]
d.clear()


print(d)
```
