```python
class Dog():
    species= "Camis familiaris"
    def __init__(self, name, age, sound):
        self.name= name
        self.age= age
        self.sound= sound


    def description(self):
        return f"{self.name} is {self.age} years old"

    def speak(self):
        return f"{self.name} says {self.sound}"

miles= Dog("Miles", 4, "bark")
miles.description()
miles.speak()


# buddy= Dog("buddy", 8)
# jack= Dog("jack", 5)

# print(miles.name)
# print(buddy.name)
# print(jack.name)



```
